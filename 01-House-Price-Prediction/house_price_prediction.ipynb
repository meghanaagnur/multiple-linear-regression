```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
```

```python
df = pd.read_csv('data\homeprice.csv', sep=",")
df
```

```python
import math
median_bedroom = math.floor(df.bedroom.median())
median_bedroom
```

```python
df.bedroom = df.bedroom.fillna(median_bedroom)
df.columns
```

```python
df
```

```python
reg = LinearRegression()
reg.fit(df[['area', 'bedroom', 'age']],df.price)
```

```python
coefficient = reg.coef_
```

```python
intercept =reg.intercept_
```

```python
reg.predict([[3000, 3, 40]])
```

```python
prediction = 3000 * coefficient[0] + 3 * coefficient[1] + 40 * coefficient[2] + intercept
print(prediction)
```

```python
reg.predict([[2500, 4, 5]])
```

```python
prediction = 2500 * coefficient[0] + 4 * coefficient[1] + 5 * coefficient[2] + intercept
print(prediction)
```

```python
%matplotlib inline
plt.xlabel('Bedroom,Age')
plt.ylabel('Price')
plt.scatter(df['bedroom'], df.price, color='blue', marker='*', label='bedroom')
plt.scatter(df['age'], df.price, color='green', marker='o', label='age')
plt.legend()
```

```python
plt.scatter(df[['area']], df.price, color='red', marker='+', label='area')
plt.legend()
```

```python
X = df[['area', 'bedroom', 'age']]
y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.4, random_state=42)
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
results = X_test.copy()
results["Actual price"]= y_test.values
results["Predicted price"]= y_pred
results
```

```python
r2= r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
print(f"R^2, {r2:.2f}")
print(f"MAE, {mae:.2f}")
print(f"MSE, {mse:.2f}")
print(f"RMSE, {rmse:.2f}")
```

```python
plt.figure(figsize=(10,4))
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Price")
plt.scatter(results["Actual price"], results["Predicted price"], color='pink', marker = '*', label="Predictions")
plt.plot([results['Actual price'].min(),results['Actual price'].max()],[results['Actual price'].min(),results['Actual price'].max()], color='black', linestyle="--", label ="Actual")
plt.legend()
```
